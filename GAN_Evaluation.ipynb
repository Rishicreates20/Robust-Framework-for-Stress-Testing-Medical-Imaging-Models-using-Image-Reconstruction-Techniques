```python
import numpy as np
from PIL import Image

# Function to calculate Diversity Score
def calculate_diversity_score(images):
    # Ensure images is a list of valid images
    if len(images) == 0:
        raise ValueError("The list of images is empty!")

    # Convert each image to a numpy array and flatten it
    images_np = []
    for image in images:
        if isinstance(image, Image.Image):  # Ensure the image is a PIL Image
            image = np.array(image)  # Convert to numpy array
        # Check if the image contains NaN values
        if np.any(np.isnan(image)):
            print("Warning: Image contains NaN values. Skipping.")
            continue
        images_np.append(image.flatten())  # Flatten the image

    # Convert the list to a numpy array
    images_np = np.array(images_np)

    # If images_np is empty after filtering, raise an error
    if images_np.shape[0] == 0:
        raise ValueError("No valid images after processing.")

    # Use pairwise distance calculation with numpy for efficiency
    distances = np.linalg.norm(images_np[:, np.newaxis] - images_np, axis=2)

    # Calculate Diversity Score (mean distance between all pairs of images)
    diversity_score = np.mean(distances)

    return diversity_score

# Example usage (assuming `generated_images` is a list of PIL Image objects):
# generated_images = [Image.open("path_to_image1"), Image.open("path_to_image2"), ...]

# Example to test we need to change our dataset on this :
generated_images = [Image.new('RGB', (64, 64)) for _ in range(5)]  # Dummy example images

# Calculate Diversity Score
try:
    diversity_score = calculate_diversity_score(generated_images)
    print(f"Diversity Score: {diversity_score:.4f}")
except ValueError as e:
    print(e)

# For Inception Score & FID, you should check that both the real and generated image arrays are valid and not empty.
```
